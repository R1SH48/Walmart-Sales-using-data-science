# This file demonstrate the sales analysis of Walmart in US

```python
import pandas
```

```python
dataset = pandas.read_csv('Walmart_Sales.csv')
```

```python
dataset.head()
```

```python
dataset['Date'] = pandas.to_datetime(dataset['Date'], format='%d-%m-%Y')
```

```python
dataset.info()
```

```python
#dataset = dataset.reset_index()
```

```python
sales_wrt_time = dataset.drop(['Holiday_Flag','Temperature','Fuel_Price','CPI','Unemployment','Store'],axis=1)
```

```python
sales_wrt_time
```

```python
grouped_sales_wrt_time = sales_wrt_time.groupby(['Date']).sum().reset_index()
```

```python
grouped_sales_wrt_time
```

# Using matplotlib 

```python

import matplotlib.pyplot as mat
```

```python
mat.figure(figsize=(25,5))
mat.plot(
    dataset.Date,
    dataset.Weekly_Sales,
    c='green'
)
```

# Using seaborn

```python
import seaborn as sns
```

```python
#fig,ax = mat.subplots(figsize=(25,10))
sns.relplot(
    data=grouped_sales_wrt_time,
    x='Date',
    y='Weekly_Sales',
    kind='line',
    height=5,
    aspect=4
)
```

# Observing the corelation of the sales with different factors

### We see that the sales see a masive spike in the month of december and january. That represent the sales are higher during the holiday season.

### We can also see the corelation with the major holidays of the year, some of which are haloween, christmas, etc. 

```python
sns.relplot(
    data=dataset,
    x='Date',
    y='Holiday_Flag',
    kind='line',
    errorbar=None,
    height=5,
    aspect=4
)
```

### The temprature tells us the sales are significantly higher during the winters

```python
sns.relplot(
    data=dataset,
    x='Date',
    y='Temperature',
    kind='line',
    errorbar=None,
    height=5,
    aspect=4
)
```

### Below are some of the factors that are related to the sales figure, but are not directly impacting the sales. Factors like : fuel price, CPI, and unemployment rate

```python
sns.relplot(
    data=dataset,
    x='Date',
    y='Fuel_Price',
    kind='line',
    errorbar=None,
    height=5,
    aspect=4
)
```

```python
sns.relplot(
    data=dataset,
    x='Date',
    y='CPI',
    kind='line',
    errorbar=None,
    height=5,
    aspect=4
)
```
